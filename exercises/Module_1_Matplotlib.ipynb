```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.xlabel('t')
plt.ylabel('t,t**2,t**3')
plt.title('Simple Plots')
plt.grid(True)
plt.axis([0, 6, 0, 150])
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--',label='t')
plt.plot(t, t**2, 'bs',label='t**2')
plt.plot(t, t**3, 'g^',label='t**3')
plt.xlabel('t')
plt.ylabel('t,t**2,t**3')
plt.title('Simple Plots')
plt.legend()
plt.grid(True)
plt.axis([0, 6, 0, 150])
```

```python
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.yscale('log')
plt.title('Log Plots')
plt.grid(True)
```

## Subplots

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(20,10))

axes[0].plot(t1, f(t1), 'bo', t2, f(t2), 'k')
axes[1].plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

plt.bar(labels,sizes)

plt.show()
```

## Pie Plot

```python
import matplotlib.pyplot as plt

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',startangle=90)

plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

plt.hist(x, bins=100, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,10))

np.random.seed(19680801)
all_data = [np.random.normal(0, std, 100) for std in range(6, 10)]

axes[0].violinplot(all_data)
axes[0].set_title('Violin plot')

# plot box plot
axes[1].boxplot(all_data)
axes[1].set_title('Box plot')
```
